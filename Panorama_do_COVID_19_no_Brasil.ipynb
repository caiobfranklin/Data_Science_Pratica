# Panorama do COVID-19 no Brasil

Na história da humanidade, já passamos por várias pandemias que causam diversas mortes.

Atualmente a população mundial tem enfrentado a pandemia causada pelo coronavirus. 

De acorco com o Ministério da Saúde, coronavírus é uma família de vírus que causam infecções respiratórias. 

O novo agente do coronavírus foi descoberto em 31/12/19, após casos registrados na China. Provoca a doença chamada de coronavírus (COVID-19).

Os primeiros coronavírus humanos foram isolados pela primeira vez em 1937. No entanto, foi em 1965 que o vírus foi descrito como coronavírus, em decorrência do perfil na microscopia, parecendo uma coroa.

A maioria das pessoas se infecta com os coronavírus comuns ao longo da vida, sendo as crianças pequenas mais propensas a se infectarem com o tipo mais comum do vírus. 

Os coronavírus mais comuns que infectam humanos são o alpha coronavírus 229E e NL63 e beta coronavírus OC43, HKU1.

A COVID-19 é uma doença causada pelo coronavírus SARS-CoV-2, que apresenta um quadro clínico que varia de infecções assintomáticas a quadros respiratórios graves. 

De acordo com a Organização Mundial de Saúde (OMS), a maioria dos pacientes com COVID-19 (cerca de 80%) podem ser assintomáticos e cerca de 20% dos casos podem requerer atendimento hospitalar por apresentarem dificuldade respiratória e desses casos aproximadamente 5% podem necessitar de suporte para o tratamento de insuficiência respiratória (suporte ventilatório).

## Obtenção dos Dados

No Brasil o orgão responsável pela divulgação oficial dos dados do coronavirus é o [Ministério da Saúde](https://coronavirus.saude.gov.br). 

Os dados são oriudos das diversas unidades de saúde de todo Brasil e percorrem um longo caminho até tornarem conhecidos.

A dificuldade de testagem, principalmente no início da pandemia por falta de insumos e teste confiáveis, dificultou a notificação dos casos confirmados.

Para este estudo, serão utlizados os dados do [Our World in Data](https://ourworldindata.org) por unificar os dados não só do Brasil, mas de outros países, para fins de comparação e análise.
 

### Dicionário de Variáveis

* iso_code = ISO 3166-1 alfa-3 - códigos de países com três letras
* continent = Continente da localização geográfica
* location = Localização geográfica
* date = Data da observação
* total_cases = Total de casos confirmados de COVID-19
* new_cases = Novos casos confirmados de COVID-19
* total_deaths = Novas mortes atribuídas ao COVID-19
* new_deaths = Novas mortes atribuídas ao COVID-19 
* total_cases_per_million = Total de casos confirmados para COVID-19 por milhão
* new_cases_per_million = Novos casos confirmados para COVID-19 por milhão
* total_deaths_per_million = Total de mortes por COVID-19 por milhão
* new_deaths_per_million = Novas de mortes por COVID-19 por milhão 
* total_tests = Total de teste para COVID-19
* new_tests = Novos testes para COVID-19
* new_tests_smoothed = Novos testes para COVID-19 (suavizado por 7 dias). Para países que não relatam dados de teste diariamente, assumimos que o teste mudou igualmente diariamente em qualquer período em que nenhum dado foi relatado. Isso produz uma série completa de números diários, que são calculados como média em uma janela de 7 dias consecutivos
* total_tests_per_thousand = Total de testes a cada 1.000 pessoas
* new_tests_per_thousand = Novos testes a cada 1.000 pessoas
* new_tests_smoothed_per_thousand = Novos testes para COVID-19 (7 dias suavizados) por 1.000 pessoas
* tests_units = Unidades usadas pelo local para relatar seus dados de teste
* stringency_index = Índice de Estresse da Resposta do Governo: medida composta baseada em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibições de viagens, redimensionada para um valor de 0 a 100 (100 = resposta mais estrita)
* population = população em 2020
* population_density = Número de pessoas divididas por área terrestre, medida em quilômetros quadrados, no ano mais recente disponível
* median_age = Idade média da população, projeção da ONU para 2020
* aged_65_older = Proporção da população com 65 anos ou mais, ano mais recente disponível
* aged_70_older = Percentagem da população com 70 anos ou mais em 2015
* gdp_per_capita = produto com paridade de poder de compra (dólares internacionais constantes de 2011), último ano disponível
* extreme_poverty = Percentagem da população que vive em extrema pobreza, último ano disponível desde 2010
* cvd_death_rate = doença cardiovascular em 2017 (número anual de mortes por 100.000 pessoas)
* diabetes_prevalence = Prevalência de diabetes (% da população entre 20 e 79 anos) em 2017
* female_smokers = Percentagem de mulheres que fumam, último ano disponível
* male_smokers = Proporção de homens que fumam, último ano disponível
* handwashing_facilities = Percentagem da população com instalações básicas de lavagem das mãos nas instalações, último ano disponível
* hospital_beds_per_thousand = Camas de hospital por 1.000 pessoas, último ano disponível desde 2010
* life_expectancy = Expectativa de vida no nascimento em 2019





Caso seja do interesse, a descrição original do nome de cada coluna pode ser encontrada [Descrição Colunas](https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data-codebook.md).

### Importando os dados do COVID-19

Os dados utilizados nesta análise foram extraídos na data de 22/07/2020 e podem ser encontrados atualizados no site [Data on COVID-19 (coronavirus) by Our World in Data](https://github.com/owid/covid-19-data/tree/master/public/data). 

Usaremos o formato `csv` neste projeto.
    

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# importar o dataset COVID-19 para um DataaFrame
df =  pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
```

#### Países com maior número de casos e mortes: 

Vamos verificar os 05 países com maior número de casos confirmado e mortes. 

```
df.loc[df.date == '2020-07-22'].sort_values(by="total_cases", ascending=False)[0:6]
```

#### Representação Gráfica

Para melhor visualiazão iremos poltar um gráfico de barras  com os 5 países com maior número de registros de morte por COVID-19

```
df_TotalMortes = df.loc[df.date == '2020-07-22', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]
```

```
df_TotalMortes
```

```
fig, ax = plt.subplots(figsize=(10,6))

df_TotalMortes.plot(x="location", y="total_deaths", kind="bar", ax=ax)
ax.set_title("Top 5 Total Deaths - (22 jul 2020)")
ax.set_xlabel("Localização")
ax.set_ylabel("Total de Mortes por COVID-19")
fig.autofmt_xdate()

plt.tight_layout()
```

Vamos também fazer uma visualização gráfica do número de mortes no mundo, para melhor vizualização será plotado um gráfico de linhas para mostrar o crescimento em relação ao número de dias.

```
df_world = df.loc[df.location == "World", "total_deaths"].reset_index(drop=True)
```

```
df_world
```

```
fig, ax = plt.subplots(figsize=(10,6))

df_world.plot(x="World", y="total_deaths", kind="line", ax=ax)
ax.set_title("Top 5 Total Deaths - (22 jul 2020)")
ax.set_xlabel("Número de dias")
ax.set_ylabel("Total de Mortes por COVID-19")


plt.tight_layout()
```

Iremos também verificar se correlação entre no número de mortes e a renda per cápita da populaçao.
Para isso iremos um scatter para melhor visualização.

```
df_2 = df.loc[df.date == '2020-07-22', ['gdp_per_capita', 'total_deaths']].sort_values(by="total_deaths", ascending=False)
```

```

# plotar o scatterplot
fig, ax = plt.subplots()

# plotar median_income vs. median_house_value
df_2.plot.scatter(x='gdp_per_capita', y='total_deaths', ax=ax, s=3)

plt.show()
```

No gráfico acima verifica-se que o número de mortes aumenta quando dimuniu o número da renda per capita.

### Análise Exploratória para o Brasil

Agora vamos realizar uma análise específica no Basil.
Para isso vamos criar um DataFrame com os dados específicos do Brasil.


```
df_brasil = df.loc[df.location == "Brazil"]
```

```
df_brasil.set_index("date", inplace=True)
```

```
df_brasil.head()
```

#### Identifcação data do primeira notificação

Vamos realizar um filtro para extrair do DataFrame a data da primeira notificação.
Depois com intenção de identificar o número de dias entre a primeira notificação mundial e a primeira notificação no Brasil, iremos contar os valores de 'total_cases=0'.

```
filter_ = df_brasil['total_cases'] >= 1.0
```

```
df_brasil[filter_].head()
```

```
df_brasil.total_cases.value_counts()
```

```
Conseguimos identificar que a primeiro notificação ocorreu em 26/02/2020, 57 dias após o primeiro relato mundial.
```

#### Representação gráfica

Iremos plotar os gráficos de linha de total de casos e total de mortes.


```
fig, ax = plt.subplots(figsize=(12,8))

ax.plot("total_deaths", data=df_brasil)
ax.plot("total_cases", data=df_brasil)
ax.set_xlabel("Data")
ax.set_ylabel("Casos")
ax.set_title("Casos confirmados x Casos de morte de Covid-19")
ax.legend()

fig.tight_layout()
```

No gráfico acima não fica muito evidente o crescimento de número de mortes em comparação ao número de casos.
Desta forma iremos plotar o gráfico em escala logarítima para ficar mais nítido o crescimento do número de mortes.

```
fig, ax = plt.subplots(figsize=(12,8))

ax.plot("total_deaths", data=df_brasil)
ax.plot("total_cases", data=df_brasil)
ax.set_xlabel("Data")
ax.set_ylabel("Casos")
ax.set_title("Casos confirmados x Casos de morte de Covid-19")
ax.legend()
ax.set_yscale("log")

fig.tight_layout()
```

## Conclusão

Apesar do Brasil ter tido quase dois meses (57 dias), período entre a primeira notificação mundial e o primeiro caso notificado no Brasil ( 26/02/2020), esse  tempo não foi suficiente para impedir e de se prepar para evitar a contaminação, sendo o segundo país com maior número de casos confirmados e o segundo com maior número de mortes, perdendo apenas para os EUA.
